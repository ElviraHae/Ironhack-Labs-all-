## Cleaning Freguesias DS

```python
# Importing the file
original_freguesias_db = pd.read_csv('Freguesias2012.csv')


# Making this col be .upper()
original_freguesias_db['FREGUESIAS53'] = original_freguesias_db['FREGUESIAS53'].str.upper()


original_freguesias_db.head(10)
```

```python
# Selecting the cols that we want to keep
freguesias_db = original_freguesias_db[['NOME', 'AREA_M2', 'FREGUESIAS53']]


# Renaming the cols
freguesias_db.columns = ['CURRENT FREGUESIAS', 'AREA', 'OLD FREGUESIAS']


# Making the Freguesias names be .upper()
freguesias_db['CURRENT FREGUESIAS'] = freguesias_db['CURRENT FREGUESIAS'].str.upper()
main_db = freguesias_db


main_db.head()
```

## Cleaning Public Lighting DS

```python
# Importing the file
original_lights_db = pd.read_excel('candeeiros.xls')


original_lights_db.head()
```

```python
# Selecting the cols we want
lights_db = original_lights_db[['OBJECTID_1', 'FREGUESIA', 'MORADA']]


# Renaming the cols
lights_db.columns = ['OBJECT NUMBER', 'CURRENT FREGUESIAS', 'PUBLIC LIGHTS']


lights_db.head(10)
```

```python
# Checking the shape

lights_db.shape
```

```python
# Counting lights for each freguesia - aka doing a 'groupby'
lights_grouped = lights_db.groupby(['CURRENT FREGUESIAS'], as_index = False).agg({'PUBLIC LIGHTS':'count'})


main_db = pd.merge(main_db, lights_grouped, on = 'CURRENT FREGUESIAS', how = 'left')


main_db.head()
```

```python
# Checking the shape again

main_db.shape
```

## Cleaning Health Centers DS

```python
# Importing the file
original_health_db = pd.read_csv('Health centers.csv')


original_health_db.head()
```

```python
# Selecting the columns with the data that we need
health_db = original_health_db[['ACTIVIDADE_NOME', 'FREGUESIA']]


# Renaming
health_db.columns = ['HEALTH CENTER', 'CURRENT FREGUESIAS']


# Making it .upper()
health_db['CURRENT FREGUESIAS'] = health_db['CURRENT FREGUESIAS'].str.upper()


health_db.head(10)
```

```python
# Doing a groupby
count_health = health_db.groupby(by = 'CURRENT FREGUESIAS', as_index = False).agg('count')


# Renaming the columns
count_health.columns = ['CURRENT FREGUESIAS', 'HEALTH CENTER']


count_health.head()


# Merging again into main database
main_db = pd.merge(main_db, count_health, on = 'CURRENT FREGUESIAS', how = 'left')

main_db.head()
```

## Merging Population into Freguesia

```python
# ELVIRA'S CODE

freguesias_db.head()

# 1: Create a dictionary for matching old Freguesias with new Freguesias with New as keys
    # and List of old as value
dict_change = freguesias_db[['CURRENT FREGUESIAS','OLD FREGUESIAS']].to_dict('list')

liste_old_Fregue = [s.split(' + ') for s in freguesias_db['OLD FREGUESIAS']]
liste_new_Fregue = freguesias_db['CURRENT FREGUESIAS']

dictionaryNEW_OLD = dict(zip(liste_new_Fregue,liste_old_Fregue))

#print(len(dictionaryNEW_OLD))


# 2: change list of values = OLD FREGUESIAS to key and add New Freguesias as value
dictionaryOLD_NEW = {} 
count = -1

for value in dictionaryNEW_OLD.values():
    count += 1
    for x in value:
        dictionaryOLD_NEW[x] = list(dictionaryNEW_OLD.keys())[count]
#print(len(dictionaryOLD_NEW))


# 3: import dictionary into a dataframe
dfOLD_NEW = pd.DataFrame.from_dict(dictionaryOLD_NEW, orient = 'index', columns = ['CURRENT\
FREGUESIAS']).reset_index()
                                                                             



# 4: add the New Freguesias to the population dataframe
pop_new = pd.merge(pop_db, dfOLD_NEW, on = 'FREGUESIAS', how = 'left')
#print(pop_new)
#print(pop_db)

# 5: group the population by the new Freguesias
pop_grouped = pop_new.groupby(['CURRENT FREGUESIAS'], as_index = False).agg({'TOTAL RESIDENTS':'sum'})


# 6: merge the population with the frequesias
main_db = pd.merge(main_db, pop_grouped, on = 'CURRENT FREGUESIAS')


main_db.head()
```
